# Soru 5:

1.sınıf, 2.sınıf ve 3.sınıf yolcuların ölüm oranlarını karşılaştırınız. Bir veya bir kaç cümle ile sonucu <b>yorumlayınız.</b>


```python
birinci= df_titanic[(df_titanic.Survived==0)&(df_titanic.Pclass==1) ].Survived.count() / df_titanic[(df_titanic.Pclass==1)].Pclass.count()
ikinci= df_titanic[(df_titanic.Survived==0)&(df_titanic.Pclass==2) ].Survived.count() / df_titanic[(df_titanic.Pclass==2)].Pclass.count()
ucuncu= df_titanic[(df_titanic.Survived==0)&(df_titanic.Pclass==3) ].Survived.count() / df_titanic[(df_titanic.Pclass==3)].Pclass.count()
display(birinci, ikinci, ucuncu)
print("Birinci sınıfın ölüm oranı 0,37, ikinci sınıfın 0,57 ve üçüncü sınıfın 0,75'tir. 1. sınıf yolcuları kurtulmaya öncelik verildiği açıkça görülmektedir.")
```

1.sınıf, 2.sınıf ve 3.sınıf bilet fiyatlarını karşılaştırınız

```python
display(df_titanic[(df_titanic.Pclass==1)].Fare.mean())
display(df_titanic[(df_titanic.Pclass==2)].Fare.mean())
display(df_titanic[(df_titanic.Pclass==3)].Fare.mean())
print("1. sınıftaki yolcular ortalama 84.15 verirken üçüncü sınıftaki yolcular 13.67 vermiştir. Büyük bir fiyat farkı vardır.")
```

# Missing Value Doldurma

Bundan sonraki soruları <b>train_missing.csv</b> datasını kullanarak cevaplayınız.

```python
eksik=pd.read_csv('train_missing.csv')
eksik
```

# Soru 8

Age, Embarked ve Fare değişkenlerini mean/median/mode değerlerini kullanarak değişken tipine uygun olacak şekilde doldurunuz.

```python
eksik1 = eksik.copy()
eksik1.Age = eksik1.Age.fillna(eksik1.Age.mean())
display(eksik1.Age)
eksik1.Embarked.mode()[0]
eksik1.Embarked = eksik1.Embarked.fillna(eksik1.Embarked.mode()[0])
display(eksik1)
```

# Soru 9

SibSp değişkeni hangi yöntemler ile doldurulabilir, fikir yürütün. Uygun bir yöntem ile doldurun. (SibSp değişkeni kişilerin gemide bulunan kardeş/eş sayısını belirtiyor.)

```python
from sklearn.impute import SimpleImputer
imp_mode = SimpleImputer(strategy='most_frequent')
imp_mode.fit(eksik[["SibSp"]])
eksik[["SibSp"]] = imp_mode.fit_transform(eksik[["SibSp"]])
display(eksik)
```

# Soru 10

Datayı baştan okuyun. Embarked değişkenini "NAN" şeklinde yeni bir kategorisi olacak şekilde doldurun.

```python
eksik["Embarked"] = eksik["Embarked"].fillna(value="NAN")

display(eksik)
```

# Soru 11

"Fare" değişkeni için "Fare_Missing" adında yeni bir değişken türetip "Fare" değişkenini median ile doldurun.

```python
eksik["Fare_Missing"] = eksik.Fare.fillna(value=df_titanic.Fare.median())
display(eksik)
```
